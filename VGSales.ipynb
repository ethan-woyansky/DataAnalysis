```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as datetime
```

```python
data = pd.read_csv('data/vgsales.csv', index_col = 'Year', parse_dates = True)
data.index = data.index.to_period('Y')
```

```python
data = data.dropna()
```

```python
plat = data.pivot_table(index = 'Platform', aggfunc = 'sum')['Global_Sales'].sort_values(ascending = False)
```

```python
%matplotlib inline
plt.title('Top 10 Consoles by Video Game Sales')
plt.xlabel('Console')
plt.ylabel('Sales (Millions)')
plt.style.use('seaborn')
plt.bar(x = plat[:10].index, height = plat[:10], color = 'Green');
```

```python
ggenre = data.pivot_table(index = 'Genre', aggfunc = 'sum')['Global_Sales'].sort_values(ascending = False)
ggenre
```

```python
%matplotlib inline
plt.title('Top 10 Video Game Genres (by sales)')
plt.xlabel('Genre')
plt.xticks(rotation = 'vertical')
plt.ylabel('Sales (Millions)')
plt.style.use('seaborn')
plt.bar(x = ggenre[:10].index, height = ggenre[:10]);
```

```python
wg = data.pivot_table('Global_Sales', ['Platform', 'Genre'], aggfunc = 'sum').loc['Wii']
wg = wg['Global_Sales'].sort_values(ascending = False)
```

```python
plt.bar(x = wg.index, height = wg, color = 'red')
plt.title('Wii Sales by Genre', fontsize = 18)
plt.xlabel('Genre')
plt.xticks(rotation = 'vertical')
plt.ylabel('Sales (millions of $)');
```

```python
%matplotlib inline
YearlySales = data.pivot_table(index = 'Year', columns = 'Genre', aggfunc = 'sum')
YearlySales = YearlySales.drop(columns = 'Rank')
YearlySales = YearlySales.dropna()
YearlySales['Global_Sales'].plot()
plt.title('Global Video Game Sales by Genre', fontsize = 18)
plt.ylabel('Sales in Millions ($)')
plt.style.use('seaborn');
```
